```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Creación de un DataFrame

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
califications = [100, 93, 72, 55, 30, 15]

cal_dict = {'years': years, 'califications': califications}

print(cal_dict)
```

## Creación de un DataFrame a partir de un diccionario

```python
cal_df = pd.DataFrame(cal_dict)

cal_df.head()
```

## Creación de visualizaciones con los datos

```python
fig = plt.figure()

plt.plot('years', 'califications', data=cal_df)

plt.title('Calificaciones a lo largo de los años')
plt.xlabel('Años')
plt.ylabel('calificación')

plt.show()
```

## Filtrado de los datos

```python
netflix_df['type']
```

```python
netflix_df_movies = netflix_df[netflix_df['type'] == 'Movie']

netflix_df_movies_columns = netflix_df[['title', 'country', 'genre', 'release_year', 'duration']]

netflix_df_movies_columns.head()
```

## Creando un scatter plot

```python
fig = plt.figure(figsize=(12,8))

plt.scatter('release_year', 'duration', data=netflix_df_movies_columns)

plt.title('Duración de las películas por año')

plt.show()
```

```python
netflix_df_movies_columns.nsmallest(n=1, columns = 'duration')
```

```python
netflix_df_movies_columns.nlargest(n=1, columns = 'duration')
```

```python
netflix_df_movies_columns['genre'].unique()
```

```python
netflix_df_movies_filtered = netflix_df_movies_columns[
    (netflix_df_movies_columns['genre'] == 'Dramas') |
    (netflix_df_movies_columns['genre'] == 'Comedies') |
    (netflix_df_movies_columns['genre'] == 'Documentaries')
]
```

```python
netflix_df_movies_filtered.head()
```

```python
colors = []

for lab, row in netflix_df_movies_filtered.iterrows():
  if row['genre'] == 'Dramas':
    colors.append('red')
  elif row['genre'] == 'Documentaries':
    colors.append('blue')
  else:
    colors.append('black')

colors[:10]
```

## Pintar un conjunto con color

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

plt.scatter('release_year', 'duration', data=netflix_df_movies_filtered, color=colors)

plt.title('Duration de Dramas/Documentales/Comedias a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Duración')

plt.show()
```

```python
count_genre_movies = netflix_df_movies_columns['genre'].value_counts()
```

```python
count_genre_movies.values
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

plt.bar(count_genre_movies.index, count_genre_movies.values)
plt.title('Cantidad de películas por genero')
plt.xlabel('Genero')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)

plt.show()
```

```python
movies_duration_year = netflix_df_movies_columns.groupby('release_year')['duration'].mean().reset_index()
movies_duration_year.head()
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

plt.plot('release_year', 'duration', data=movies_duration_year)
plt.title('Duración de las películas por año')
plt.xlabel('Año')
plt.ylabel('Duración')

plt.show()
```

```python
netflix_movies_country = netflix_df_movies_columns['country'].value_counts().head(10)
netflix_movies_country.head()
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,5))

plt.barh(netflix_movies_country.index, netflix_movies_country.values)
plt.title('Cantidad de películas por país')
plt.xlabel('Cantidad')
plt.ylabel('País')

plt.show()
```
